```python
from sklearn import datasets, linear_model, model_selection
import matplotlib.pyplot as plt

# Load the diabetes datasets
X, y = datasets.load_diabetes(return_X_y=True)
```

```python
# Select only the blood pressure feature
X = X[:, 3]
print(X[:5])

# Reshape into 2D
X = X.reshape((-1, 1))
print(X[:5])
```

```python
# Split the dataset into training and test set
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.3)

# Train the model
model = linear_model.LinearRegression()
model.fit(X_train, y_train)
```

```python
# Make predictions using tbe test dataset
y_pred = model.predict(X_test)
```

```python
# Create a plot to visualize the predictions
plt.scatter(X_test, y_test, color="black")
plt.plot(X_test, y_pred, color="blue")
plt.xlabel("Average blood pressure")
plt.ylabel("Disease progression")
plt.title("A Graph Plot Showing Disease progression against Average blood pressure")
plt.show()
```
