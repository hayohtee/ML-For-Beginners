```python
import pandas as pd

# Read the US-pumpkins CSV file
pumpkins = pd.read_csv("../data/US-pumpkins.csv")
pumpkins = pumpkins[pumpkins['Package'].str.contains('bushel', case=True, regex=True)]
pumpkins.head()
```

```python
# Check for missing data
pumpkins.isnull().sum()
```

```python
# Select the essential columns
columns = ["Package", "Low Price", "High Price", "Date"]
pumpkins = pumpkins.loc[:, columns]
```

```python
# Calculate the average price of pumpkin
price = (pumpkins["Low Price"] + pumpkins["High Price"]) / 2
month = pd.DatetimeIndex(pumpkins["Date"]).month
```

```python
# Create new DataFrame based on the extracted features
new_pumpkins = pd.DataFrame({
    "Month": month, "Package": pumpkins["Package"],
    "Price": price, "Low Price": pumpkins["Low Price"],
    "High Price": pumpkins["High Price"]
})

new_pumpkins.head()
```

```python
# Update the price of pumpkins based on the package
new_pumpkins.loc[new_pumpkins["Package"].str.contains("1 1/9"), "Price"] = price/(1 + 1/9)
new_pumpkins.loc[new_pumpkins["Package"].str.contains("1/2"), "Price"] = price/(1/2)
new_pumpkins.head()
```

## Visualizing the data

```python
import matplotlib.pyplot as plt

price = new_pumpkins.loc[:, "Price"]
month = new_pumpkins.loc[:, "Month"]
plt.scatter(month, price)
plt.xlabel("Month")
plt.ylabel("Price per bushel")
plt.show()
```

```python
# Making it useful
new_pumpkins.groupby(["Month"])["Price"].mean().plot(kind="bar")
plt.ylabel("Pumpkin price per bushel")
plt.show()
```
